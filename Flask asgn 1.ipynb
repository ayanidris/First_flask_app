In this example, we import the Flask class and the url_for function from the flask module. We then create a new Flask object and define three routes using the @app.route decorator:

The first route ('/') returns a simple "Hello, World!" message.
The second route ('/user/<username>') expects a username parameter in the URL and uses it to generate a personalized welcome message.
The third route ('/post/<int:post_id>') expects an integer post_id parameter in the URL and returns a message with the post number.
Finally, we use the with app.test_request_context() context manager to create a test request context and call the url_for() function with different arguments to generate URLs for the defined routes. We print the generated URLs to the console.

```python
from flask import Flask, url_for

app = Flask(_name_)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Welcome to your profile, {username}!'

@app.route('/post/<int:post_id>')
def show_post(post_id):
    return f'Post {post_id}'

with app.test_request_context():
    print(url_for('index'))          # Output: '/'
    print(url_for('user_profile', username='John Doe'))  # Output: '/user/John%20Doe'
    print(url_for('show_post', post_id=123))  # Output: '/post/123'
```
